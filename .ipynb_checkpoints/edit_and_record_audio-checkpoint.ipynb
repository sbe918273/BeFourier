```python
import pyaudio
import matplotlib.pyplot as plt
import numpy as np
from numpy import cos, pi
import time
import BeFourier as bf
%matplotlib notebook
```

```python
audio = pyaudio.PyAudio()

FORMAT = pyaudio.paInt16 # We use 16bit format per sample
CHANNELS = 1
CHUNK = 1024 # 1024 bytes of data red from a buffer
RATE = 8000

stream_in = audio.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True)#,
                    #frames_per_buffer=CHUNK)

def get_audio_data(record_seconds=1, rate=44100, scale=1, as_bytes=True):
    
    raw_data = stream_in.read(int(record_seconds * rate))
    scaled_encoded_data = np.frombuffer(raw_data, np.int16) * scale

    return scaled_encoded_data.tobytes() if as_bytes else scaled_encoded_data
    
stream_out = audio.open(format=FORMAT,
                   channels=CHANNELS,
                   rate=RATE,
                   output=True)
```

```python
stream_in.start_stream()
print("NOW")
g = get_audio_data(record_seconds=2, rate=RATE, as_bytes=False)
print("STOP")
stream_in.stop_stream()
```

```python
stream_out.write((g * 20).tobytes())
```

```python
g_len=len(g)
t = np.array([i/RATE for i in range(int(-0.5*g_len), int(0.5*g_len))])
```

```python
fig, ax = plt.subplots(figsize=(9,4))
ax.plot(t, g)
```

```python
nu = np.linspace(0, 500, 500)
F = [2*bf.F_func(g, t, i) for i in nu]
```

```python
fig, axes = plt.subplots(figsize=(9,4))
axes.plot(nu, [z.mod for z in F])
```

```python
g_hat = [bf.F_func(F, nu, i) for i in -1*t]
audio_out = np.array([z.re for z in g_hat], dtype='int16')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
axes[0].plot(t, audio_out)
axes[1].plot(t, g)
```

```python
stream_out.write((audio_out *20).tobytes())
```

<hr/>

```python
fig, ax = plt.subplots(figsize=(9,4))
F_flattened = bf.flatten(F, nu, 400, 500)
ax.plot(nu, [z.mod for z in F_flattened])
```

```python
g_desire_hat = [bf.F_func(F_flattened, nu, i) for i in -1*t]
desire_audio_out = np.array([z.re for z in g_desire_hat], dtype='int16')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
axes[0].plot(t, desire_audio_out)
axes[1].plot(t, g)
```

```python
stream_out.write((desire_audio_out *60)[:int(len(g)/2)].tobytes())
```

```python
stream_out.write((g*60).tobytes())
stream_out.write((desire_audio_out *60)[:int(len(g)/2)].tobytes())
```
