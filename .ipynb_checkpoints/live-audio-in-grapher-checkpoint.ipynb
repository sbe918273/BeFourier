```python
import pyaudio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import wave
%matplotlib notebook
```

```python
audio = pyaudio.PyAudio()

FORMAT = pyaudio.paInt16 # We use 16bit format per sample
CHANNELS = 1
RATE = 44100
CHUNK = 1024 # 1024bytes of data red from a buffer
RECORD_SECONDS = 0.1
WAVE_OUTPUT_FILENAME = "file.wav"

stream = audio.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True)#,
                    #frames_per_buffer=CHUNK)
stream.start_stream()

get_audio_data = lambda: np.frombuffer(stream.read(CHUNK), np.int16)
```

```python
fig, ax = plt.subplots(figsize=(6,3))

ax.set_ylim(-40, 40)
t = np.linspace(0, 1024, 1024)
line, = ax.plot(t, get_audio_data())

def animate(count):
    
    line.set_data(t, get_audio_data())
    

    
anim = FuncAnimation(fig, animate, frames=10000, interval=1)

```

```python
stream.stop_stream()
stream.close()

audio.terminate()
```
